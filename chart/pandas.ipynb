```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams["font.family"] = "Apple SD Gothic Neo"
plt.rcParams["axes.unicode_minus"] = False

import warnings
warnings.simplefilter("ignore") # 무시할게
```

```python
df = pd.DataFrame(np.random.randn(1000,3), columns=list('ABC'))
df = df.cumsum()

df
```

```python
# 범례가 포함된 선 차트
g = df.plot()
```

```python
g = df.plot(legend=False)   # 범례 뺌
```

```python
# g = df["B"].plot()
g = df.plot(y="B")
```

```python
# 보조축 설정

df["A"].plot()
df["B"].plot(secondary_y=True,style="r")
```

```python
# 보조축 + 레이블 지정

# ax = df.plot(secondary_y=["A","B"])   범례에 right 표시
ax = df.plot(secondary_y=["A","B"],mark_right=False)
ax.set_ylabel("C_Scale")
ax.right_ax.set_ylabel("AB_Scale")
plt.show()
```

```python
# 컬럼별 선 차트 따로 그리기

g = df.plot(subplots=True)
```

```python
np.random.seed(1000)

df = pd.DataFrame(np.random.randn(1000,4), columns=list('ABCD'),index=pd.date_range('1/1/2021',periods=1000))

df
```

```python
# 2021-01-06 날짜로 바 차트 작성
# iloc : 숫자, loc : 레이블명
# g = df.iloc[5].plot.bar()
g = df.iloc[5].plot(kind="bar")
```

```python
df2 = pd.DataFrame(np.random.rand(50,4), columns=list('abcd'))
```

```python
df2.plot.scatter(x="a",y="b")
```

```python
# s : 점 크기 설정

g = df2.plot.scatter(x="a",y="b",s=50)
```

```python
ax = df2.plot.scatter(x="a",y="b",color="DarkBlue",label="Group 1")
df2.plot.scatter(x="c",y="d",color="DarkGreen",label="Group 2",ax=ax)
```

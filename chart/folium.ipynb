```python
# 마커 + 아이콘

m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=12
)

# 마커지정
tooltip = "Click me!!"

folium.Marker(
    # 돈의문 박물관
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
    icon=folium.Icon(color='red',icon='info-sign'),
    tooltip=tooltip
).add_to(m)

folium.Marker(
    [37.5693255,126.9860066],
    popup="<i>솔데스크</i>",
    icon=folium.Icon(color='blue',icon='star'),
    tooltip=tooltip
).add_to(m)

folium.Marker(
    [37.5716352,126.9767435],
    popup="<i>광화문광장</i>",
    icon=folium.Icon(color='green',icon='bookmark'),
    tooltip=tooltip
).add_to(m)

m
```

```python
# BeautifyIcon 으로 아이콘을 만들수있다

m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=12
)

icon_plane = plugins.BeautifyIcon(
    icon="plane", border_color='darkblue', text_color='darkblue', icon_shape='circle'
)

folium.Marker(
    [37.562402,126.801289],
    popup="<i>김포공항</i>",
    icon=icon_plane
).add_to(m)


folium.Marker(
    # 돈의문 박물관
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
    icon=folium.Icon(color='red',icon='info-sign'),
).add_to(m)

folium.Marker(
    [37.5693255,126.9860066],
    popup="<i>솔데스크</i>",
    icon=folium.Icon(color='blue',icon='star'),
).add_to(m)

folium.Marker(
    [37.5716352,126.9767435],
    popup="<i>광화문광장</i>",
    icon=folium.Icon(color='green',icon='bookmark'),
).add_to(m)

m
```

```python
m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=12
)

folium.Marker(
    # 돈의문 박물관
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
).add_to(m)

m.add_child(folium.ClickForMarker(popup="Marker"))  # 클릭할때 마다 아이콘이 생김 , 팝업에 마커라는 글자를 보여줌
```

```python
m = folium.Map(
    location=[37.566697,126.978426],
    zoom_start=15
)

folium.Circle(  # 마커를 동그라미로 표시할 수 있음
    # 돈의문 박물관
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
    radius=60,
    color="royalblue",
    fill=True,  # 동그라미 안에 안채움(False), 채움(True)
    fill_color="darkblue"
).add_to(m)

m
```

```python
cities = [
    [37.566687, 126.978417],
    [37.179774, 129.075004],
    [37.455900, 128.705522],
    [36.350451, 127.384827],
    [35.160072, 126.851440],
]

m = folium.Map(
    location=[37.566697,126.978426],
    zoom_start=7
)

for i in range(len(cities)):
    folium.Circle(
        location=cities[i],
        radius=200,
        color="red",
    ).add_to(m)

m
```

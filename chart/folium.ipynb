# pip install vincent
    - Vega() 차트 표시할 때 차트 작성해주는 라이브러리

```python
import vincent
import json

scatter_points = {  # 임의의 숫자
    "x" : np.random.randn(50).cumsum(),
    "y" : np.random.randn(50).cumsum(),
}

# 
scatter_chart = vincent.Scatter(scatter_points,iter_idx="x", width=400, height=200) # 점 찍는 차트(스캐터)
scatter_json = scatter_chart.to_json()  # json 의 형태로 만들었음
scatter_dict = json.loads(scatter_json) # json을 딕셔너리 구조로 만들었음

m = folium.Map([36.5,128],zoom_start=7)

# Vega() : 지도 팝업 안에 차트 표시
popup = folium.Popup()
folium.Vega(scatter_json,height=250, width=450).add_to(popup)
folium.Marker([37,128],popup=popup).add_to(m)

popup = folium.Popup()
folium.Vega(scatter_dict,height=250, width=450).add_to(popup)
folium.Marker([36.5,127],popup=popup).add_to(m)

m
```

```python
url = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"

# 미국 위,경도 => 다운로드 함
# state_geo = requests.get(f"{url}/us-states.json") 이 코드 사용 안함

# 실업률 정보
state_unemployment = f"{url}/US_Unemployment_Oct2012.csv"

state_data = pd.read_csv(state_unemployment)
state_data.head()
```

```python
from branca.colormap import linear

colormap = linear.YlGnBu_09.scale(  # state_data 의 Unemployment 데이터를 색상으로 나타내겠음
    state_data.Unemployment.min(),
    state_data.Unemployment.max()
)

colormap
```

```python
# 지도 색상별로 실업율 표시
# folium.GeoJson + branca

state_data_dict = state_data.set_index("State")["Unemployment"] # state_data 인덱스 설정해서 딕셔너리 구조로 바꿈

m = folium.Map([43,-100],zoom_start=4)

folium.GeoJson(
    "./us-states.json",
    name="Unemployment",
    style_function=lambda feature:{
        "fillColor":colormap(state_data_dict[feature["id"]]),   # 데이터를 id 를 찾아서 매칭함
        "color":"black",
        "weight":1,
        "dashArray":"5.5",
        "fillOpacity":0.6
    }
).add_to(m)

folium.LayerControl().add_to(m)
m
```

```python
# 지도 색상별로 실업율 표시
# folium.Choropleth

m = folium.Map([43,-100],zoom_start=4)

folium.Choropleth(
    geo_data='./us-states.json',
    data=state_data,
    columns=["State","Unemployment"],
    key_on="feature.id",
    fill_color="YlGnBu",
    fill_opacity=0.6,
    line_opacity=0.2,
    legend_name="Unemployment Rate(%)"
).add_to(m)

folium.LayerControl().add_to(m)
m
```

```python
url = "https://raw.githubusercontent.com/suanlab/dataset/master"

seoul_geo_url =f"{url}/seoul_municipalities_geo_simple.json"
seoul_geo = requests.get(seoul_geo_url).json()  # 위,경도 정보가 json으로 들어가 있음
seoul_geo
```

```python
seoul_population = f"{url}/seoul_population.csv"
seoul_data = pd.read_csv(seoul_population,encoding="utf-8")
seoul_data.head()
```

```python
m = folium.Map([37.528043,126.980238],zoom_start=10)

folium.GeoJson(
    seoul_geo,
    name="seoul_municipalities"
).add_to(m)

m
```

```python
# 인구수에 맞는 컬러바 생성

colormap = linear.Blues_09.scale(
    seoul_data.population.min(),
    seoul_data.population.max()
)

colormap
```

```python
# 구명 인덱스로 설정
population_dict = seoul_data.set_index("name")["population"]
population_dict.head()
```

```python
# 구 별 컬러 지정

color_dict = {str(key):colormap(population_dict[key]) for key in population_dict.keys()}    # 딕셔너리 구조
color_dict
```

```python
#  구별 인구수 색상별로 표현

m = folium.Map([37.528043,126.980238],zoom_start=10)

folium.GeoJson(
    seoul_geo,
    name="population",
    style_function=lambda feature:{
        "fillColor":color_dict[feature["properties"]["name"]],  # 컬러맵을 딕셔너리로 미리 만들었음 - 딕셔너리 구조라 [] 대괄호로 불러옴 - 2차원배열이라서 이렇게 불러옴
        "color":"black",
        "weight":1,
        "dashArray":"5,5",
        "fillOpacity":0.6
    }
).add_to(m)

colormap.add_to(m)

folium.LayerControl().add_to(m)
m
```

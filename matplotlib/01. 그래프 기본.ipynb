```python
import matplotlib.pyplot as plt
```

### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
list1 = [10,20,30,40]

# 차트 함수 : plot()
plt.plot(list1)
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

g = plt.plot(x,y)   # 변수로 받으면 matplotlib 객체 정보 어쩌구가 안 뜬다
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict1 = {"data_x":[1,2,3,4,5],"data_y":[2,3,5,7,9]}

plt.plot("data_x","data_y",data=dict1)
plt.show()
```

### 타이틀 설정

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title("Chart")
g = plt.plot(x,y)
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title("선차트")
g = plt.plot(x,y)
```

### 마이너스 설정

```python
x = [1,2,3,4]
y = [12,-5,25,15]

plt.title("선차트")
g = plt.plot(x,y)
```

```python
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 한글설정 / 마이너스 설정 필수
plt.rcParams["font.family"] = "MesloLGS NF"
plt.rcParams["axes.unicode_minus"] = False
```

### 그리드

**전체설정**

```python
plt.plot(x,y)
plt.grid(True)
```

**x축설정**

```python
plt.plot(x,y)
plt.grid(True,axis="x")
```

**y축설정**

```python
plt.plot(x,y)
plt.grid(True,axis="y")
```

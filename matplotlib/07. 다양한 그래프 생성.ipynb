```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Apple SD Gothic Neo"
plt.rcParams["axes.unicode_minus"] = False
```

### 1) 막대차트(기본) - bar()

```python
labels = ["강백호","서태웅","정대만"]
values = [175,187,184]

g = plt.bar(labels, values)
```

><b>색상</b>

```python
labels = ["강백호","서태웅","정대만"]
values = [175,187,184]

g = plt.bar(labels, values, color="r")
```

```python
labels = ["강백호","서태웅","정대만"]
values = [175,187,184]
colors = ["r","g","b"]

g = plt.bar(labels, values, color=colors)
```

```python
labels = ["강백호","서태웅","정대만"]
values = [175,187,184]
colors = ["r","g","b"]

g = plt.bar(labels, values, color=colors, alpha=0.5)
```

><b>축 레이블 조정</b>

```python
labels = ["강백호","서태웅","정대만"]
values = [175,187,184]
colors = ["r","g","b"]

plt.ylim(175,190)   # limit 을 조정하면 특정 데이터를 나오지않게 할 수 있음(ticks 조절 혹은 limit 조절)
g = plt.bar(labels, values, color=colors)
```

><b>막대 너비 조정</b>

```python
g = plt.bar(labels, values, color=colors, width=0.5)
```

><b>축 레이블 rotation 설정</b>

```python
plt.xticks(rotation=45) # 레이블 기울여서 화면 출력
plt.yticks(rotation=45)
g = plt.bar(labels, values, color=colors, width=0.5)
```

><b>막대 </b>

```python
# x축 강백호, 서태웅, 정대만 ==> 임의의 레이블 지정
xticks = ["1번학생","2번학생","3번학생"]

plt.bar(labels, values, color=colors, width=0.5)
plt.xticks(labels,xticks)
plt.show()
```

### 2) 막대차트(심화) - barh()

**수평막대**

```python
g = plt.barh(labels,values)
```

```python
plt.xlim(170,195)
g = plt.barh(labels,values)
```

**막대 패턴**(https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

```python
g = plt.barh(labels,values)
g[0].set_hatch("/")
g[1].set_hatch("x")
g[2].set_hatch("o")
```

```python
g = plt.bar(labels,values)

# get_height() : 막대 높이 가져오기
for idx,b in enumerate(g):
    plt.text(idx,b.get_height()+0.5, values[idx], ha="center")
```

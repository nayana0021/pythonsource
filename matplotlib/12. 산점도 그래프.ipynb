### 산점도 그래프(scatter)

- x, y 상관관계를 나타낼 때 사용

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams["font.family"] = "Apple SD Gothic Neo"
plt.rcParams["axes.unicode_minus"] = False
```

```python
df1 = pd.read_csv("./student.csv")
df1
```

#### 학년 컬럼 추가

```python
# 3,3,2,1,1,3,2,2
df1["학년"] = [3,3,2,1,1,3,2,2]
df1
```

#### 기본 scatter

```python
# 영어, 수학 점수는 서로 관계가 있느냐?

plt.scatter(df1["영어"],df1["수학"])
plt.xlabel("영어")
plt.ylabel("수학")
plt.show()
```

#### 원 크기 변경 - 랜덤 값 이용

```python
# rand() : 0~1 사이의 균일분포 난수를 생성
# rand(8) : 0~1 사이의 균일분포 난수 8개를 생성

np.random.rand(8)
```

```python
np.random.seed(42)  # 실행 시 동일값 발생시키려고
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df1["영어"],df1["수학"],s=sizes)
plt.xlabel("영어")
plt.ylabel("수학")
plt.show()
```

#### 원 크기 변경 - 학년 값 이용

```python
sizes = df1["학년"] * 500
plt.scatter(df1["영어"],df1["수학"],s=sizes)
plt.xlabel("영어")
plt.ylabel("수학")
plt.show()
```

#### 색상 변경

#### cmap (https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
plt.scatter(df1["영어"],df1["수학"],s=sizes,cmap="viridis",c=df1["학년"])
plt.xlabel("영어")
plt.ylabel("수학")
plt.show()
```

#### alpha : 투명

```python
plt.scatter(df1["영어"],df1["수학"],s=sizes,cmap="viridis",c=df1["학년"],alpha=0.3)
plt.xlabel("영어")
plt.ylabel("수학")
plt.show()
```

#### colorbar() 

```python
plt.scatter(df1["영어"],df1["수학"],s=sizes,cmap="viridis",c=df1["학년"],alpha=0.3)
plt.xlabel("영어")
plt.ylabel("수학")
plt.colorbar()
plt.show()
```

```python
plt.scatter(df1["영어"],df1["수학"],s=sizes,cmap="viridis",c=df1["학년"],alpha=0.3)
plt.xlabel("영어")
plt.ylabel("수학")
plt.colorbar(ticks=[1,2,3],label="학년",shrink=0.8,orientation="horizontal")
plt.show()
```

- ticks : 기본으로 나오는 눈금 대신 다른 것 지정
- label : 컬러바 라벨 지정
- shrink : 컬러바 길이 지정
- horizontal : 컬러바 위치(가로)

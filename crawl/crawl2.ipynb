```python
# get 방식

import requests

# 세션을 이용한 요청 가능
s = requests.Session()

r = s.get("https://www.naver.com")

print(r.status_code)
print(r.ok) # ok냐 True 가 넘어오고있음 / if문을 활용한 코드 작성 가능

print(r.text)

# 세션 종료
s.close()
```

```python
# get / post / put / delete 함수 제공함 - 연습할수있는 사이트에서 해봐야함

import requests

s = requests.Session()

r = s.get("https://httpbin.org/get")

print(r.text)

# 세션 종료
s.close()
```

```python
# get / post / put / delete 함수 제공함 - 연습할수있는 사이트에서 해봐야함

import requests

s = requests.Session()

# post + data
data = {
    "name":"hong"
}

r = s.post("https://httpbin.org/post",data=data)

print(r.text)

# 세션 종료
s.close()
```

```python
# get / post / put / delete 함수 제공함 - 연습할수있는 사이트에서 해봐야함

import requests

s = requests.Session()

# delete + data
data = {
    "name":"hong"
}

r = s.delete("https://httpbin.org/delete",data=data)

print(r.text)

# 세션 종료
s.close()
```

```python
# get / post / put / delete 함수 제공함 - 연습할수있는 사이트에서 해봐야함

import requests

s = requests.Session()

# put + data
data = {
    "name":"hong"
}

r = s.put("https://httpbin.org/put",data=data)

print(r.text)

# 세션 종료
s.close()
```

```python
# get + parameter ?name=hong&age=14

import requests

s = requests.Session()

param = {"name":"hong", "age":14}

r = s.get("https://httpbin.org/get",params=param)

print(r.text)

# 세션 종료
s.close()
```

```python
import requests

r = requests.get("https://jsonplaceholder.typicode.com/todos/1")
# print(r.text) 텍스트
print(r.headers)    # 헤더 정보를 가져오고 싶을때
print(r.encoding)
print(r.json())
print(r.json().keys())
print(r.json().values())
```

```python
r = requests.get("https://jsonplaceholder.typicode.com/users")

for row in r.json():
    for k,v in row.items():
        print(k,":",v)
    print()
```

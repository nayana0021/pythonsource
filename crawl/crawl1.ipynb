```python
import requests

urls = ["https://www.python.org/","https://www.naver.com/"]

name = "robots.txt"

for url in urls:
    print(url+name)
    robots = requests.get(url+name)
    print(robots.text)
```

```python
import urllib.request as req

url = "https://google.com"

save_file = "./google.html"

try:
    file1, header1 = req.urlretrieve(url,save_file)
except Exception as e:
    print(e)
else:
    print(header1)
    print(file1)
    print("성공")
```

#### urlretrieve
    - 요청하는 url의 정보를 로컬 파일로 저장
    - 튜플로 반환됨
    - csv, api 데이터 등 많은 양의 데이터를 한번에 저장할 때 사용

```python
img_url = "https://search.pstatic.net/sunny/?src=https%3A%2F%2Fi.pinimg.com%2Foriginals%2F07%2F23%2F39%2F07233970f93ea3cebdb3cfaa2b773253.jpg&type=sc960_832"
html_url = "https://www.naver.com"

# 경로
img_path = "./cat.jpg"
html_path = "./naver.html"

try:
    file1, header1 = req.urlretrieve(img_url, img_path)
    file2, header2 = req.urlretrieve(html_url, html_path)
except Exception as e:
    print(e)
else:
    print("file1 {}".format(file1))
    print("file2 {}".format(file2))
```

#### 다운로드는 필요없고 내용만 분석하고 싶을 때
    - urlopen(), read()

```python
weather_url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

data = req.urlopen(weather_url).read()

text = data.decode("utf-8")
print(text)
```

```python
url = "https://www.11st.co.kr/browsing/BestSeller.tmall?method=getBestSellerMain&xfrom=main^gnb"

try:
    res = req.urlopen(url)
    contents = res.read().decode('euc-kr')
except Exception as e:
    print(e)
else:
    print(res.info())
    print(contents[:4000])
```

```python
from urllib.parse import urlencode  # 파이썬에 인코딩 라이브러리 있음

url="https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&"    # 뒤에 쿼리스트링이 한글인 경우 인코딩이 되어야함

param = {"query":"영화"}    # key, value 값으로 

url += urlencode(param)

try:
    res = req.urlopen(url)
    contents = res.read().decode("utf-8")
except Exception as e:
    print(e)
else:
    print(res.info())
    print(contents[:4000])
```
